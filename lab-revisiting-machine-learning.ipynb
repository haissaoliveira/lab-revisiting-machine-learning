```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\haiss\Downloads\learningSet.txt")
data
```

#### Check for null values in all the columns

```python
null_percent = pd.DataFrame(data.isna().sum()/len(data)).reset_index()
null_percent.columns = ['column_name', 'nans_percentage']
null_percent.sort_values(by = ['nans_percentage'], ascending = False)
```

#### Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:

OSOURCE - symbol definitions not provided, too many categories

ZIP CODE - we are including state already

```python
drop_list = []
drop_list.append('OSOURCE')
drop_list.append('ZIP')
data.drop(columns=drop_list, inplace=True)
```

#### Identify columns that over 85% missing values

```python
threshold = 0.85

condition = null_percent['nans_percentage'] > threshold
higher_nan = null_percent[condition]
higher_nan
```

#### Remove those columns from the dataframe

```python
drop_list = list(higher_nan['column_name'])
data = data.drop(columns = drop_list)
```

```python
data
```

#### Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
data['GENDER']
```

```python
data['GENDER'] = data['GENDER'].fillna('F')
data['GENDER'] = data['GENDER'].apply(lambda x: x if x in ['M', 'F'] else 'other')
```
